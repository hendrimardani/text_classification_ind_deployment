```python
stop_words = nltk.corpus.stopwords.words('indonesian')
np.random.choice(stop_words, size=10)
```

```python
text = '. pokok nya kurang kurang afdol kalau ke lembang atau ke bandung tidak mampir ke sini . 4 humas naik buat kampung daun .'
nltk_tokens = nltk.word_tokenize(text)
print(nltk_tokens)
```

```python
def preprocessing(data, feature):
    for i, value in tqdm(enumerate(data[feature]), desc="Please Wait...!!!"):
        sentence = value
        sentences = re.sub(r'[^\w\s]', '', sentence)

        words = nltk.word_tokenize(sentences)
        words = [w.lower() for w in words if not w in stop_words]
        final_sentences = " ".join(words)
        data.loc[i, "text"] = final_sentences
```

```python
preprocessing(train, "text")
preprocessing(val, "text")

train
```

```python
# Predict data
# Name of the TFLite text classification model.
_MODEL = '/kaggle/working/model_text_classification_ind/model.tflite'
# Whether to run the model on EdgeTPU.
_ENABLE_EDGETPU = False
# Number of CPU threads to run the model.
_NUM_THREADS = 4

# Initialize the text classification model.
base_options = core.BaseOptions(file_name=_MODEL, use_coral=_ENABLE_EDGETPU, num_threads=_NUM_THREADS)
options = text.NLClassifierOptions(base_options)

# Create NLClassifier from options.
classifier = text.NLClassifier.create_from_options(options)
```

```python
predict_label = []
for x in range(10):
    sentence = val['text'].iloc[x]
    label = val['labels'].iloc[x]
    text_classification_result = classifier.classify(sentence)
    classification_list = text_classification_result.classifications[0].categories

    # Sort output by probability descending.
    predict_label.append(sorted(
      classification_list, key=lambda item: item.score, reverse=True)[0])

[x for x in predict_label]
```
